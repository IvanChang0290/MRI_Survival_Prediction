```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

date_column = ['Brats20ID','Age','Survival_days','Extent_of_Resection']
surInfoFile = "C:/Users/a0981/OneDrive//archive/BraTS2020_training_data/content/data/survival_info.csv"
df_surInfo = pd.read_csv(surInfoFile,names= date_column,skiprows=1)
print(df_surInfo)


plt.hist(df_surInfo['Age'], bins=10, edgecolor='black')  # Adjust bins as needed
plt.xlabel('Age')
plt.ylabel('numbers')
plt.title('Distribution of Age')
plt.show()

plt.hist(df_surInfo['Survival_days'], bins=10, edgecolor='black')  # Adjust bins as needed
plt.xlabel('days')
plt.ylabel('numbers')
plt.title('Distribution of Survival_days')
plt.show()


plt.scatter(df_surInfo['Age'], df_surInfo['Survival_days'])
plt.xlabel('Age')
plt.ylabel('Survival_days')
plt.title('Relationship between Age and Survival_days')
plt.grid(True)
plt.show()
```
